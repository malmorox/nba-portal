```python
# Celda 1: Imports
from nba_api.stats.static import teams
import pandas as pd
import matplotlib.pyplot as plt

# Celda 2: Obtener equipos
teams_data = teams.get_teams()
df = pd.DataFrame(teams_data)
df.head()

# Celda 3: Ver columnas disponibles
df.columns

# Celda 4: Ver info básica
print(f"Total equipos: {len(df)}")
df.info()

# Celda 5: Equipos por ciudad
df['city'].value_counts()

# Celda 6: Gráfico rápido
df['city'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 ciudades con equipos NBA')
plt.show()

# Celda 7: Filtrar un equipo
lakers = df[df['full_name'].str.contains('Lakers')]
lakers
```
